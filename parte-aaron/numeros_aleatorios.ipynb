### Ejercicio 1

Realizar un código básico de un generador lineal congruente que acepte como parámetros los valores a, c, m y la semilla. Modifique los valores para obtener distintos comportamientos

```python

def generador_lineal_congruente(a, c, m, seed, n):

    x = seed
    numeros = []
    for _ in range(n+1):
        x = (a * x + c) % m
        numeros.append(x)

    return numeros
```

```python
a = 11
c = 3
m = 10
seed = 42
n = 5000

numeros_aleatorios = generador_lineal_congruente(a, c, m, seed, n)
```

### Ejercicio 2

Representar los valores en un histograma y su homogeneidad en el espacio

```python
import matplotlib.pyplot as plt

plt.hist(numeros_aleatorios, bins=100, align='left', edgecolor='black')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de Números Aleatorios Generados')
plt.show()
```
