# Ecuaciones diferenciales ordinarias (EDOs)

# Ejercicio 1

**Resuelve la siguiente ODE de primer orden:**

$$
\frac{dy}{dt} = -y
$$

**Condiciones iniciales:**
$ y = 1 $, $ t = 0 $

**Intervalo:**
$ 0 \leq t \leq 3 $

**Objetivo:**
Representa $ y = y(t) $.


```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def dydt(y,t):
    return -y

y0 = 1
t = np.linspace(0,3,100)

sol = odeint(dydt, y0, t)

plt.plot(t, sol)
plt.xlabel('t')
plt.ylabel('y(t)')
```

# Ejercicio 2

Resuelve la siguiente ODE de primer orden:

$$
\frac{dx}{dt} = -x^3 + \sin t
$$

con las condiciones iniciales: $t = 0$, $x = 0$, en el intervalo $0 \leq t \leq 10$. Representa $x = x(t)$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def dxdt(x, t):
    return -x**3 + np.sin(t)

t = np.linspace(0,10, 100)
x0 = 0
sol = odeint(dxdt, x0, t)

plt.plot(t, sol)
plt.xlabel('t')
plt.ylabel('x(t)')
```

### Ejercicio 3

Resuelve el siguiente sistema de ODEs de primer orden:

$$
\begin{align*}
\frac{dx}{dt} &= x - xy - 0.1x^2, \\
\frac{dy}{dt} &= xy - y - 0.05y^2,
\end{align*}
$$

con las condiciones iniciales: $t = 0$, $x = 2$, $y = 1$, en el intervalo $0 \leq t \leq 30$. Representar $x = x(t)$, $y = y(t)$, $y = y(x)$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def sist(x,t):
    x, y = x
    dxdt = x-x*y -0.1*x**2
    dydt = x*y - y - 0.05*y**2
    return dxdt , dydt

t = np.linspace(0,30, 300)
x0, y0 = 2,1
sol = odeint(sist, [x0,y0], t)

xsol, ysol = sol.T

fig = plt.figure(figsize=(10,3))

ax1, ax2, ax3 = fig.add_subplot(1,3,1), fig.add_subplot(1,3,2), fig.add_subplot(1,3,3)

ax1.plot(t, xsol)
ax1.set_xlabel('t')
ax1.set_ylabel('x(t)')

ax2.plot(t, ysol)
ax2.set_xlabel('t')
ax2.set_ylabel('y(t)')

ax3.plot(xsol, ysol)
ax3.set_xlabel('x')
ax3.set_ylabel('y(x)')

fig.tight_layout()
```

# Ejercicio 4

Resuelve el siguiente sistema de ODEs de primer orden

$$
\frac{dx}{dt} = -0.1x
$$

$$
\frac{dy}{dt} = 0.1x - 0.2y
$$

con las condiciones iniciales: $t = 0$, $x = 100$, $y = 0$, en el intervalo $0 \leq t \leq 20$. Representar $x = x(t)$, $y = y(t)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def sist(x,t):
    x, y = x
    dxdt = -0.1*x
    dydt = 0.1*x - 0.2*y
    return dxdt , dydt

t = np.linspace(0,20, 300)
x0, y0 = 100,0
sol = odeint(sist, [x0,y0], t)

xsol, ysol = sol.T

fig = plt.figure(figsize=(10,4))
ax1, ax2 = fig.add_subplot(1,2,1), fig.add_subplot(1,2,2) 

ax1.plot(t, xsol)
ax1.set_xlabel('t')
ax1.set_ylabel('x(t)')

ax2.plot(t, ysol)
ax2.set_xlabel('t')
ax2.set_ylabel('y(t)')

fig.tight_layout()
plt.show()
```
