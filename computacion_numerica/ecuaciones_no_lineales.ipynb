# Ecuaciones no lineales

Resolver $$x+b\cos x -a = 0$$ con $a = 4, b = 2, -5<x<5$


```python
from scipy.optimize import fsolve
import numpy as np

def f(x):
    return x + 2*np.cos(x) - 4

x0 = 4

sol = fsolve(f, x0)

print(sol[0])
```

# Ejericio 2

La magnetización de un material viene dada por:

$$ m = \tanh \frac{Cm}{T} $$

Si $C=1$, representar $m$ en $0K<T<2K$

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

def magnetizacion(m, T):
    return m - np.tanh(m/T)

T = np.linspace(0.01,2,100)

m = []
for t in T:
    x0 = 0.5
    sol = fsolve(magnetizacion, x0, args=(t))[0]
    m.append(sol)
plt.plot(T,m)
plt.xlabel('T')
plt.ylabel('m')
plt.title('Magnetizacion en función de T')
plt.show()
```
