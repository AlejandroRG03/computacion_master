# Ecuaciones no lineales

Resolver $$x+b\cos x -a = 0$$ con $a = 4, b = 2, -5<x<5$


```python
from scipy.optimize import fsolve
import numpy as np

def f(x):
    return x + 2*np.cos(x) - 4

x0 = 4

sol = fsolve(f, x0)

print(sol[0])
```

# Ejercicio 2

La magnetizaciÃ³n de un material viene dada por:

$$ m = \tanh \frac{Cm}{T} $$

Si $C=1$, representar $m$ en $0K<T<2K$

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

def magnetizacion(m, T):
    return m - np.tanh(m/T)

T = np.linspace(0.01,2,100)

m = []
for t in T:
    x0 = 0.5
    sol = fsolve(magnetizacion, x0, args=(t))[0]
    m.append(sol)
plt.plot(T,m)
plt.xlabel('T')
plt.ylabel('m')
plt.title('Magnetizacion en funciÃ³n de T')
plt.show()
```

Resolver la ecuaciÃ³n
$$0.9 x^4 - 4x^3 + 0.5 x^2 + 11x -4.5 = 0$$

Con $-2<x<4$

```python
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 0.9*x**4 - 4*x**3 + 0.5*x**2 + 11*x -4.5
x = np.linspace(-2,4,100)
plt.plot(x, f(x))
plt.plot([-2,4],[0,0])

# Polinomimo de grado 4 -> 4 soluciones complejas

sols = [fsolve(f,-2), fsolve(f,0), fsolve(f,2), fsolve(f,4)]

for sol in sols:
    print('x = ', sol[0])#:) 
```
