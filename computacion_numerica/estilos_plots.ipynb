# Ejercicio 1

Crear una colección ordenada constituída por $n=15$ valores de $x_i$, equiespaciados entre 1 y 3. A partir de ella crear otra lista de 15 valores dada por $y_i=x_i^2e^{-x_i^2}$.

Representar $y(x)$ y poner puntos representados como círculos rojos y enrejado

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(1,3,15)
y = x**2 * np.exp(-x**2)

plt.plot(x,y,'ro')
plt.grid(linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y(x) = x^2e^{-x^2}$')

#plt.tick_params(axis = 'x', length = 9, width = 0.4, labelsize = 10) #Personalizar ejes

plt.show()

```

# Ejercicio 2

Sean las calificaciones almacenadas en la lista:
$$ \text{calfs} = [8.5, 7.0, 5.5, 9.0, 4.0, 10.0, 9.5, 7.5] $$

El programa deberá obtener y mostrar en pantalla:

- El promedio de las calificaciones
- La calificación más alta
- La calificación más baja
- La cantidad y el porcentaje de aprobados
- La cantidad y el porcentaje de suspensos

**Atención**

1. Definir una función para cada uno de los puntos anteriores
2. Mostrar los resultados en la pantalla con `print`

```python
def promedio(lista):
    suma = 0
    len = 0
    for elem in lista:
        suma = suma + elem
        len = len + 1
    return suma/len

def maximo(lista):

    maximo = lista[0]

    for elem in lista[1:]:
        if elem > maximo:
            maximo = elem
    return maximo

def minimo(lista):

    minimo = lista[0]

    for elem in lista[1:]:
        if elem < minimo:
            minimo = elem
    return minimo

def porcentaje_aprobados(lista):

    aprobados = 0
    len = 0

    for elem in lista:
        if elem >= 5:
            aprobados = aprobados + 1
        len = len + 1
    return aprobados/len*100

def porcentaje_suspensos(lista):
    return 100 - porcentaje_aprobados(lista)

calfs = [8.5, 7.0, 5.5, 9.0, 4.0, 10.0, 9.5, 7.5]

print(f"El promedio de calificaciones es: {promedio(calfs)}")
print(f"La calificación máxima es: {maximo(calfs)}")
print(f"La calificación mínima es: {minimo(calfs)}")
print(f"El porcentaje de aprobados es: {porcentaje_aprobados(calfs)} %")
print(f"El porcentaje de suspensos es: {porcentaje_suspensos(calfs)} %")
```

Escribir el código para obtener cierto subplot

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(9,6))

#fig.subplots_adjust(wspace=0.5,hspace=0.5)

ax1 = fig.add_subplot(3,2,1); ax2 = fig.add_subplot(3,2,2); ax3 = fig.add_subplot(3,2,(3,6))

ax1.set_xlabel('x'); ax2.set_xlabel('x'); ax3.set_xlabel('x')
ax1.set_ylabel('y'); ax2.set_ylabel('y'); ax3.set_ylabel('y')

plt.suptitle('add_subplot')

ax1.set_title('add_subplot(3,2,1)',size=10); ax2.set_title('add_subplot(3,2,2)',size=10); ax3.set_title('add_subplot(3,2,(3,6))',size=10)

ax1.set_xlim([0,3]); ax2.set_xlim([0,3]); ax3.set_xlim([-2,2])
ax1.set_yticks(np.arange(0,1,0.2)); ax2.set_yticks(np.arange(0,1,0.2))

fig.tight_layout() # Esto ajusta automaticamente para que no solapen las figuras

plt.show()
```
