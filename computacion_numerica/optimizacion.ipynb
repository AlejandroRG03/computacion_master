# Ejercicio 2

**Dada la función:**

$$
y = -\frac{\text{sen } x}{x} - 5
$$

**Hallar todos los mínimos contenidos en el intervalo**
$-15 \leq x \leq 15$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

def f(x):
    
    return np.where(x != 0, -np.sin(x)/x -5, -6)

xp = np.linspace(-15,15,100)
plt.plot(xp, f(xp))
plt.xlabel('x')
plt.ylabel('y')

xtry = np.arange(-15,15, 1)

mins = [np.round(fmin(f, x0=xtry[0], disp=False)[0], 2)]

for i in range(1,len(xtry)):

    min = np.round(fmin(f, x0=xtry[i], disp=False)[0], 2)

    if min not in mins:

        mins.append(min)

mins = np.array(mins)
fmins = f(mins)
print('Los mínimos son:')
for i in range(len(mins)):
    
    print(f'x = {mins[i]}, y = {fmins[i]:.2f}')
```

## Ejercicio 3

Dada la función:

$$
y = -\frac{\sin x}{x} - 5
$$

Hallar el **mínimo global** en el intervalo $-15 \leq x \leq 15$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

def f(x):
    
    return np.where(x != 0, -np.sin(x)/x -5, -6)
xp = np.linspace(-15,15,100)
plt.plot(xp, f(xp))
plt.xlabel('x')
plt.ylabel('y')

xtry = np.arange(-15,15, 1)

mins = [np.round(fmin(f, x0=xtry[0], disp=False)[0], 2)]

for i in range(1,len(xtry)):

    min = np.round(fmin(f, x0=xtry[i], disp=False)[0], 2)

    if min not in mins:

        mins.append(min)

mins = np.array(mins)
fmins = f(mins)

print(f'El mínimo global es x = {mins[np.argmin(fmins)]}, y = {np.min(fmins)}')
```

## Ejercicio 4

Dada la función:

$$
y = -\frac{\sin x}{x} - 5
$$

Hallar **todos los máximos y mínimos** contenidos en el intervalo $-15 \leq x \leq 15$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

def f(x):
    return np.where(x != 0, -np.sin(x)/x -5, -6)
def g(x):
    return -f(x)


xp = np.linspace(-15,15,100)
plt.plot(xp, f(xp))
plt.xlabel('x')
plt.ylabel('y')

xtry = np.arange(-13,13, 1)

mins = [np.round(fmin(f, x0=xtry[0], disp=False)[0], 2)]
maxs = [np.round(fmin(g, x0=xtry[0], disp=False)[0], 2)]

for i in range(1,len(xtry)):

    min = np.round(fmin(f, x0=xtry[i], disp=False)[0], 2)
    max = np.round(fmin(g, x0=xtry[i], disp=False)[0], 2)
    if min not in mins:
        mins.append(min)
    if max not in maxs:
        maxs.append(max)

mins = np.array(mins)
maxs = np.array(maxs)
fmins = f(mins)
fmaxs = f(maxs)

print('Mínimos:')
for i in range(len(mins)):
    print(f'x = {mins[i]}, y = {fmins[i]:.2f}')

print('Máximos:')
for i in range(len(maxs)):
    print(f'x = {maxs[i]}, y = {fmaxs[i]:.2f}')
```

## Ejercicio 5

Calcular el **mínimo** de la función:

$$
f(x, y) = x^2 + y^4 + x + y
$$

para el dominio:
$$
-5 \leq x \leq 5 \\
-5 \leq y \leq 5


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

def f(x):
    x, y = x
    return x**2 + y**4 + x + y

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = y = np.linspace(-5,5,100)
ax.plot(x,y, f([x,y]))


minimo = fmin(f, x0=(0,0), disp=False)

print(f'El mínimo es x = {minimo[0]:.2f}, y = {minimo[1]:.2f}')
```

## Ejercicio 7

Calcular el **mínimo** de la función:

$$
f(x, y) = x e^{-(x^2 + y^2)}
$$

para el dominio:
$$
-1 \leq x \leq 1 \\
-1 \leq y \leq 1


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

def f(x):
    x, y = x
    return x*np.exp(-x**2-y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = y = np.linspace(-1,1,100)
ax.plot(x,y, f([x,y]))


minimo = fmin(f, x0=(0,0), disp=False)

print(f'El mínimo es x = {minimo[0]:.2f}, y = {minimo[1]:.2f}')
```
